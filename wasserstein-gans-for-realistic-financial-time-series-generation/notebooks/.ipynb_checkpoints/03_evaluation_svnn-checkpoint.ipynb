## 1. Configuration & Data

```python
from quantgan import DataConfig, PreprocessConfig, DatasetConfig, ModelConfig, TrainConfig
from quantgan.data import DefeatBetaSource, LambertWPreprocessor
from quantgan.evaluation import PaperEvaluator, Plotter
from quantgan.utils import set_all_seeds, build_and_load_generator, generate_M_paths_raw

data_cfg = DataConfig()
pre_cfg  = PreprocessConfig()
ds_cfg   = DatasetConfig()
model_cfg = ModelConfig(
    generator_type="svnn",
    g_ch=60,
    g_ch_hidden=60,
)
train_cfg = TrainConfig()

WEIGHTS_PATH = "./quantgan_outputs_wgangp/bestG_svnn_seed0_epoch220.weights.h5"
M      = 500
Ttilde = 4000
S_lags = 250
SEED   = 0

set_all_seeds(SEED)
```

```python
src = DefeatBetaSource(data_cfg)
df = src.fetch()
close = df["Close"].dropna()
real_series = src.log_returns_from_close(df)
T_real = len(real_series)

print(f"[DATA] T={T_real} mean={float(real_series.mean()):.6g} std={float(real_series.std()):.6g}")
```

```python
pre = LambertWPreprocessor(pre_cfg).fit(real_series)
print("[PRE]", pre.summary())
```

## 2. Load Model

```python
netG = build_and_load_generator(
    model_cfg=model_cfg,
    window_len=ds_cfg.window_len,
    weights_path=WEIGHTS_PATH,
    seed=SEED,
)
print(f"[MODEL] {model_cfg.generator_type} | params: {netG.count_params():,}")
```

## 3. Debug: Raw Generator Output

```python
import tensorflow as tf
_z = tf.random.normal([5, Ttilde + ds_cfg.window_len - 1, model_cfg.z_dim], seed=0)
_raw = netG(_z, training=False).numpy()[..., 0]
print(f"raw output  mean={_raw.mean():.4g}  std={_raw.std():.4g}  min={_raw.min():.4g}  max={_raw.max():.4g}")
print(f"pre u_std={pre.state.u_std:.4g}  u_mean={pre.state.u_mean:.4g}")
```

## 4. Generate Paths

```python
fake_paths = generate_M_paths_raw(
    netG=netG,
    preproc=pre,
    M=M,
    Ttilde=Ttilde,
    window_len=ds_cfg.window_len,
    z_dim=model_cfg.z_dim,
    batch=50,
    seed=SEED,
)
print(f"[GEN] shape={fake_paths.shape}")
print(f"[GEN] mean={float(fake_paths.mean()):.6g}  std={float(fake_paths.std()):.6g}")
print(f"[REAL] mean={float(real_series.mean()):.6g}  std={float(real_series.std()):.6g}")
```

## 5. Paper Score

```python
evaluator = PaperEvaluator(
    real_series=real_series,
    preproc=pre,
    train_cfg=train_cfg,
    dy_base_t=len(real_series),
)
paper_score, parts = evaluator.paper_score(fake_paths)

print(f"=== PAPER SCORE: {paper_score:.6g} ===")
print(f"ACF(r):   {parts['acf_x']:.4g}")
print(f"ACF(|r|): {parts['acf_abs']:.4g}")
print(f"ACF(r²):  {parts['acf_sq']:.4g}")
print(f"Leverage: {parts['lev']:.4g}")
print(f"DY sum:   {parts['dy_sum']:.4g}")
print(f"DY by lag: {parts['dy_by_t']}")
```

## 6. Plots

```python
plotter = Plotter(show=True, save=False)
fake_for_plots = fake_paths[:, :T_real]
```

```python
plotter.plot_acf_bundle(
    real_r=real_series,
    fake_paths=fake_for_plots,
    S=S_lags,
    title_suffix=" | svnn",
)
```

```python
plotter.plot_leverage(
    real_r=real_series,
    fake_paths=fake_for_plots,
    S=S_lags,
    title_suffix=" | svnn",
)
```

```python
plotter.plot_hist_panel(
    real_r=real_series,
    fake_paths=fake_for_plots,
    t_lags=(1, 5, 20, 100),
    title="Distribution Comparison | svnn",
    label_fake="Generated",
    label_real="Historical",
)
```
